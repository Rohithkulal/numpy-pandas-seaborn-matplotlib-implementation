# mean

```python
data['Profit'] = data['Close'] - data['Open']
mean = data['Profit'].mean()
print('mean:',mean)
```

# median

```python
median = data['Profit'].median()
print('median:',median)
```

# Standard Deviation

```python
std = data['Profit'].std()
print('Standard Deviation:',std)
```

# Maximum

```python
max = data['Profit'].max()
print('Maximum:',max)
```

# Minimum

```python
min = data['Profit'].min()
print('Minimum:',min)
```

# Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


print("\nTask 5: Data Visualization")
# Visualization 1: Line plot showing the trend in sales over time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Revenue'], label='Revenue')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Trend in Sales Over Time")
plt.legend()
plt.show()

# Visualization 2: Bar plot showing the total sales for each product category
plt.figure(figsize=(10, 6))
sns.barplot(x='High', y='Revenue', data=data, estimator=sum, ci=None)
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.title("Total Sales for Each Product Category")
plt.xticks(rotation=45)
plt.show()
```
