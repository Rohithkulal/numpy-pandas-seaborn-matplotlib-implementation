# Matplotlib

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.plot(x, y,marker='o' ,color='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.plot(x, y,"k*", label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sample Line Plot')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.bar(x, y ,color='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar graph Plot')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.hist(x ,color='red', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histgram Plot')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.scatter(x, y ,color='blue', label='Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scattering Plot')
plt.legend()

plt.show()
```
